### Main Concepts in Data Visualization
How to tell the right story...

**Types of plots:**
* Histograms - show distributions
* Boxplots - show the range and skewness of values
* Bar charts - show how different categories compare
* Line plots - show how one (dependent) variable varies with respect an independent variable (e.g. over time)
* Pie charts - show relative sizes between parts of a whole
* Simgle numbers

**How to know when and what to plot?** <br />
Choosing the right plot is a matter of **intuition**. The goal is to **present the message clearly**, or "tell the right story". There are two main kinds of visualizations.
* For scientific analysis and work - **stricter rules**
    * For exploratory analysis / quick references
* For presenting the results to non-specialists - we can be creative, but we have to keep our message in mind.
* The result may also be printed or viewed on a dashboard.

Each plot has two spatial dimensions, but we can add more using color, size, even animation. To choose the right plot, think about numberical or categorical variables, structure (spatial, temporal, etc.), clustering, relative size and etc.

**How can you know what all of these are?** <br />
Perform an **exploratory data analysis** first. Play around with the data and document everithing, so you remember it later.

The **basic rules** are pretty simple. First of all, choose apropriate chart type. *If you cant, just compare.* Make the plot big enough to fit the plotting area. When it's not obvious, add a title and a legent. **Label the axes.** Point at interesting data. Use marker size and color to convey information. Don't strain the reader!

The plot should tell a story. It should be clear what it want's to say. It should make use of temporal data, **it has labaled axes** and conveys a clear message. If it is not a scientific plot, it should be interesting for the reader. It should tell as much information as possible, while still giving the information and telling the story. The good plot should also allow comparison of different distributions. It should use a clear color map and possibly not comic sans as a font. Labels also make it easier to read. Good plot is easy on the eyes and tells the whole story, not just a part of it. Stacked area charts, or transparent ones also give us a chance to see the difference easier. Legend placement is also something to note. It shouldn't hide a part of the plot. It should not have highly skewed and disproportional elements. **DO NOT F\*\*\*\*\*G USE 3D CHARTS** Perspective often lies! The chart should be easily understandable. The values should be related. And the customizations should make sence... It should show the real purpose, and HAVE AXES!! The background should not distract the reader. There should be relation between different variables on the chart, else there will be no sense at all. The data should be right. Do not make spelling errors xd...

### Exploratory Data Analysis (EDA)
Making sense of our data

Other data ploting libraries are [`Seaborn`](https://seaborn.pydata.org) and [`Bokeh`](https://docs.bokeh.org/en/latest/index.html). You could check them out. They make some pretty plots pretty easily, but have in mind that they are not as flexible as matplotlib. [`Vega-Altair`](https://altair-viz.github.io), [`GGPlot2`](https://ggplot2.tidyverse.org) and [`Plotly`](https://plotly.com/python/) are also cool options. But we need to focus on representing the data, not the actual library we use. :)

* Exploratory Data Analysis is a process to see what the data can tell us.
    * Not tied to formal data modelling or hypothesis testing
* Many people have written about this (most notably John Tukey, 1961)
* Like data cleaning, it relies heavily on the scientist's intuition.
* Objectives:
    * Suggest hypotheses
    * Assess assumptions on which models will be based
    * Aid selection of features (feature engineering)
    * Provide a basis for further data collections

Analytic Graphs are good for explainign the dataset visually. We could show distributions, relations, comparisons and casuality even in multivariate data.

+ Principles of analytic (scientific) graphs
    + Show **comparisons**
    + Show **casuality**
        + **Correlation does not imply causation**
    + Show **multivariatie data** (many variables)
    + Integrate evidence from **multiple sources**
    + Describe and **document** evidence
    + **"Content is king"**
        + We have to have something interesting to report
        
These principles apply to EDA as well
